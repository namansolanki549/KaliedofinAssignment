```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Exercise Data.csv")
df
```

```python
for i in df['Amount returned to everyone in the group']: 
    df['Net amount recd by Bid winner'] += i
    df["Total Return"] = df['Net amount recd by Bid winner']
df["Annualized Return"] = df["Total Return"] * (12/25)
df["Return Percentage"] = df["Total Return"] * (100 / 50000)
df
```

```python
print('Annualized  Return of the person who bids in the last month is: {}'.format(list(df['Annualized Return'])[-1]))
print('Annualized Return of the person who bids in the first month is: {}'.format(list(df['Annualized Return'])[0]))
```
